```python
sum_column = death_df["Conflict and terrorism"] + death_df["Terrorism"]
death_df["Terror"] = sum_column
```

                ,          

```python
death_df=death_df.drop(['Code', 'executions'],axis=1)
```

        ,         

```python
death_df.drop(death_df[death_df['Entity'] == 'World'].index , inplace=True)
death_df.drop(death_df[death_df.Entity == 'Western Europe'].index , inplace=True)
death_df.drop(death_df[death_df.Entity == 'East Asia'].index , inplace=True)
death_df.drop(death_df[death_df.Entity == 'Southeast Asia'].index , inplace=True)
```

```python
df_new =death_df.sum(axis=0).reset_index()
df_new=df_new.rename(columns={'index': 'cause', 0:'total deaths'})
df_new.drop(df_new.index[[0,1 , -1]],inplace=True)
df_new.convert_dtypes()
df_new = df_new.sort_values(by=["total deaths"],ascending=False)
```

```python
fig = px.bar(df_new, x='cause', y='total deaths',
             hover_data=['cause', 'total deaths'], color='cause',
             labels={'deaths':'causes of deaths'},width=1200, height=1200, title="Top Death causes in the World")
fig.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)
fig.show()
```
